Part 2: Several other models are used in this part and compared by their accuracy

```python
train_features_2 = pd.read_csv(open("train_features.tsv"),sep = '\t')
train_labels_2 = pd.read_csv(open("train_labels.tsv"),sep = '\t')

X_2 = train_features_2.drop(columns=['YTId','movieId'])
X_2
```

```python
from sklearn.compose import ColumnTransformer
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.preprocessing import OneHotEncoder
column_trans = ColumnTransformer(
     [('tag_category', CountVectorizer(),'tag'),
      ('title_bow', CountVectorizer(), 'title')],
     remainder='passthrough')


X_2_clean = column_trans.fit_transform(X_2)
y_2_clean = np.array(train_labels_2['genres'])
```

```python
test_features_2 = pd.read_csv(open("valid_features.tsv"),sep = '\t')
test_labels_2 = pd.read_csv(open("valid_labels.tsv"),sep = '\t')

X_test_2 = test_features_2.drop(columns=['YTId','movieId'])
X_test_2_clean = column_trans.transform(X_test_2)
y_test_2_clean = np.array(test_labels_2['genres'])
```

```python
from sklearn.svm import SVC
from sklearn.neighbors import KNeighborsClassifier
from sklearn.ensemble import AdaBoostClassifier, RandomForestClassifier
from sklearn import tree
from sklearn.linear_model import LogisticRegression
models = [
    ('SVM', SVC()),
    ('KNN', KNeighborsClassifier()),
    ('AB', AdaBoostClassifier()),
    ('RF', RandomForestClassifier()),
    ('DT', tree.DecisionTreeClassifier()),
    ('LR', LogisticRegression(random_state=0)),
    
]
```

```python
for name, model in models:
    clf = model
    clf.fit(X_2_clean, y_2_clean)
    accuracy = clf.score(X_test_2_clean, y_test_2_clean)
    print(name, accuracy)
```
